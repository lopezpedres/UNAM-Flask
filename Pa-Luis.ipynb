```python
#ORIGINAL(2020)
def url_name(h2,h5):
    #name_1 = re.findall(':(.+)',h2)[0].strip()
    name_1 = re.findall('Concurso (.+)',h2)[0].strip()
    name_2 = re.findall('-(.+)',h2)[0].strip()
    name_3 = h5.split()[0] + '-' + h5.split()[-2]
    return name_1 + "-" +name_2 + "-" + name_3
```

```python
#NEW FOR 2019
def url_name(h2,h5):
    #name_1 = re.findall(':(.+)',h2)[0].strip()
    name_1 = re.findall('Concurso (.+)',h2)[0].strip()
    name_2 = re.findall('\n(.+)',h2)[0].strip()
    name_3 = h5.split()[0] + '-' + h5.split()[-2]
    return name_1 + "-" +name_2 + "-" + name_3
```
